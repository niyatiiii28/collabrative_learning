```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
speech_conv = pd.read_csv("/content/speech_conversation_features_normalized_2.csv")
speech_pair = pd.read_csv("/content/speech_pairwise_mean_1.csv")
eye = pd.read_excel("/content/pairwise_featurematrix_eye.xlsx")
emotion = pd.read_csv("/content/emotion_pair_mean.csv")
hand = pd.read_csv("/content/pairlevel_features_hand.csv")
```

```python
hand.head()
```

```python
eye.rename(columns={"Pair_id":"pair_id"}, inplace=True)
```

```python
df = speech_conv.merge(speech_pair, on="pair_id", how="inner")
df = df.merge(eye, on="pair_id", how="inner")
df = df.merge(emotion, on="pair_id", how="inner")
df = df.merge(hand, on="pair_id", how="inner")
```

```python
pair_ids = df["pair_id"]

X = df.drop(columns=["pair_id"])
X = X.select_dtypes(include=[np.number])
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_scaled_df = pd.DataFrame(X_scaled, columns=X.columns)
X_scaled_df["pair_id"] = pair_ids.values
```

```python
X_scaled_df.head()
```

```python
pca = PCA(n_components=0.90)   # keep enough components to explain 90% variance
pca_components = pca.fit_transform(X_scaled)

pca_df = pd.DataFrame(pca_components,
                      columns=[f"PC{i+1}" for i in range(pca_components.shape[1])])
pca_df["pair_id"] = pair_ids.values

pca_df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# -------------------------------------------------------
# 1. PCA with automatic selection (retain 90% variance)
# -------------------------------------------------------
pca = PCA(n_components=0.90)
pca_components = pca.fit_transform(X_scaled)

pc_count = pca.n_components_
print("Number of components explaining â‰¥ 90% variance:", pc_count)

# Create PCA dataframe
pca_df = pd.DataFrame(
    pca_components,
    columns=[f"PC{i+1}" for i in range(pc_count)]
)
pca_df["pair_id"] = pair_ids.values

pca_df.head()
```

#Variance Explained per Component (Bar Plot)

Each bar represents how much information (variance) each principal component captures from the original dataset.

If PC1 is much higher than PC2 and others, it means PC1 captures the strongest pattern in the data.

If the bars drop slowly, it means the dataset is high-dimensional and no single PC dominates.

```python
plt.figure(figsize=(7,4))
plt.bar(range(1, pc_count+1), pca.explained_variance_ratio_)
plt.xlabel("Principal Component")
plt.ylabel("Variance Explained")
plt.title("Variance Explained by Each PCA Component")
plt.show()
```

#Cumulative Variance Plot

This plot shows how many components are needed to explain a certain amount of the total information.

For example, if the curve crosses 0.90 at PC4, it means:
â€œ4 components capture 90% of the information in the dataset.â€

```python
cum_var = pca.explained_variance_ratio_.cumsum()

plt.figure(figsize=(7,4))
plt.plot(range(1, pc_count+1), cum_var, marker='o')
plt.xlabel("Number of Components")
plt.ylabel("Cumulative Variance Explained")
plt.title("Cumulative Variance Explained by PCA")
plt.grid(True)
plt.show()

print("Cumulative Variance:", cum_var)
```

#PCA Loadings Table
The numerical version of the heatmap.
It helps you see exactly which features:

This is important for telling what each PC represents.

```python
loadings = pd.DataFrame(
    pca.components_.T,
    columns=[f"PC{i+1}" for i in range(pc_count)],
    index=X.columns
)

loadings
```

```python
for i in range(pc_count):
    print(f"\nTop features for PC{i+1}")
    print(loadings.iloc[:, i].abs().sort_values(ascending=False).head(10))
```

#PCA Loadings Heatmap

This tells you which behavioral dimensions drive the variance.

```python
plt.figure(figsize=(10,8))
sns.heatmap(loadings, cmap="coolwarm", center=0)
plt.title("PCA Component Loadings")
plt.show()
```

#PCA Scatter Plot (PC1 vs PC2)
This is the most important visualization for clustering.

Each point = one pair (P01, P02, â€¦).
Distance between points = similarity in high-dimensional space.

Points close together â†’ pairs behave similarly

Points far apart â†’ pairs behave differently

Clusters appearing visually = natural grouping

If you color by KMeans clusters, it gives you an intuitive sense of the groups.

```python
if pc_count >= 2:
    plt.figure(figsize=(7,5))
    plt.scatter(pca_df["PC1"], pca_df["PC2"], s=100)

    for i, txt in enumerate(pca_df["pair_id"]):
        plt.annotate(txt, (pca_df["PC1"][i], pca_df["PC2"][i]))

    plt.xlabel("PC1")
    plt.ylabel("PC2")
    plt.title("PCA Scatter Plot (First Two Components)")
    plt.show()
```

#K-Means

Elbow Method to Choose k

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

inertias = []
K = range(2, 10)

for k in K:
    km = KMeans(n_clusters=k, random_state=42)
    km.fit(pca_components)
    inertias.append(km.inertia_)

plt.figure(figsize=(7,5))
plt.plot(K, inertias, marker='o')
plt.xlabel("k (Number of clusters)")
plt.ylabel("Inertia")
plt.title("Elbow Method")
plt.grid(True)
plt.show()
```

Silhouette Score for Each k

```python
from sklearn.metrics import silhouette_score

sil_scores = []

for k in K:
    km = KMeans(n_clusters=k, random_state=42)
    labels = km.fit_predict(pca_components)
    score = silhouette_score(pca_components, labels)
    sil_scores.append(score)

plt.figure(figsize=(7,5))
plt.plot(K, sil_scores, marker='o')
plt.xlabel("k")
plt.ylabel("Silhouette Score")
plt.title("Silhouette Method")
plt.grid(True)
plt.show()

sil_scores
```

1. Elbow Method Interpretation

The elbow plot decreases like this:

k=2 â†’ 320

k=3 â†’ 255

k=4 â†’ 220

k=5 â†’ 160 â† big drop

k=6 â†’ 130

k=7 â†’ 98

k=8 â†’ 75

k=9 â†’ 65

Where is the elbow?
The largest drop occurs between k=4 â†’ k=5, and after that the curve bends and becomes smoother.

This visually suggests k = 5 or k = 3 as possible elbow points.

2. Silhouette Score Interpretation

Silhouette Scores:

k	silhouette
2	0.208
3	0.235 (highest)
4	0.203
5	0.232
6	0.222
7	0.233
8	0.205
9	0.176

Highest silhouette = k = 3

Second highest (almost tied) = k = 7 and k = 5

Silhouette prefers fewer clusters because your dataset is small (only 5 pairs).
With such small sample size, clustering beyond 5 clusters becomes meaningless.

So silhouette tells us:

â†’ k=3 is the best
â†’ k=5 or k=7 are acceptable alternatives

 Final Recommendation (Best Choice)
âœ” k = 3 clusters

Why?

It has the highest silhouette score

It makes the most sense given your dataset size

It balances elbow + silhouette

It avoids overfitting (creating clusters smaller than sample size)

3 groups is interpretable for pair-wise analysis

```python
k = 3
kmeans = KMeans(n_clusters=k, random_state=42)
labels = kmeans.fit_predict(pca_components)

pca_df["cluster"] = labels
pca_df
```

```python
plt.figure(figsize=(8,6))
plt.scatter(pca_df["PC1"], pca_df["PC2"], c=pca_df["cluster"], cmap="viridis", s=150)

for i, name in enumerate(pca_df["pair_id"]):
    plt.annotate(name, (pca_df["PC1"][i], pca_df["PC2"][i]), fontsize=12)

plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("PCA Clustering (k = 3)")
plt.colorbar(label="Cluster")
plt.show()
```

### Training a Simple Machine Learning Model

This example demonstrates a basic machine learning workflow using a synthetic dataset. We'll:

1.  **Generate Synthetic Data**: Create a simple dataset with two features and two classes.
2.  **Split Data**: Divide the dataset into training and testing sets.
3.  **Train a Model**: Use `LogisticRegression` from `scikit-learn` to train a classifier.
4.  **Make Predictions**: Predict class labels on the test set.
5.  **Evaluate Model**: Calculate accuracy and display a classification report.

```python
import numpy as np
import pandas as pd
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report

# 1. Generate Synthetic Data
X, y = make_classification(n_samples=100, n_features=2, n_informative=2, n_redundant=0, n_clusters_per_class=1, random_state=42)
df_synthetic = pd.DataFrame(X, columns=['Feature1', 'Feature2'])
df_synthetic['Target'] = y

print("First 5 rows of synthetic data:")
display(df_synthetic.head())

# 2. Split Data into Training and Testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print(f"\nTraining set size: {X_train.shape[0]} samples")
print(f"Testing set size: {X_test.shape[0]} samples")
```

```python
# 3. Train a Simple Model (Logistic Regression)
model = LogisticRegression(random_state=42)
model.fit(X_train, y_train)

print("Model training complete.")
```

```python
# 4. Make Predictions on the Test Set
y_pred = model.predict(X_test)

print("First 10 predictions:", y_pred[:10])
print("First 10 actual labels:", y_test[:10])
```

```python
# 5. Evaluate the Model
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)

print(f"\nModel Accuracy: {accuracy:.2f}")
print("\nClassification Report:")
print(report)
```
